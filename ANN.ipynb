# Artificial Neural Network for Customer Churn Prediction

This notebook demonstrates the use of modular Python scripts for data preprocessing, model building, and training for a customer churn prediction task. The original notebook has been refactored for better organization and maintainability.

## Project Structure

- `data_preprocessing.py`: Handles data loading, cleaning, feature engineering, and scaling.
- `model.py`: Defines the ANN architecture and compilation.
- `train.py`: Orchestrates the training process, including data splitting, class weight calculation, and model fitting.
- `ANN.ipynb`: This notebook, which now calls functions from the modular scripts.
- `README.md`: Project overview and setup instructions.

## Setup and Installation

1.  **Clone the repository:**
    ```bash
    git clone https://github.com/sar-adnan/ANN-on-Churn-Product-Prediction.git
    cd ANN-on-Churn-Product-Prediction
    ```

2.  **Create a virtual environment and install dependencies:**
    ```bash
    python -m venv venv
    source venv/bin/activate  # On Windows, use `venv\Scripts\activate`
    pip install -r requirements.txt
    ```

3.  **Run the pipeline:**
    You can run the training pipeline directly from the `train.py` script:
    ```bash
    python train.py
    ```
    Or, execute the cells below in this notebook.
